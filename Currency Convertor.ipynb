```python
import tkinter as tk
from tkinter import ttk

# Define exchange rates (as of a specific date)
exchange_rates = {
    'USD': 1.0,
    'EUR': 0.85,
    'GBP': 0.73,
    'JPY': 110.13,
    'INR': 74.77,
}

def convert_currency():
    try:
        amount = float(entry_amount.get())
        from_currency = combo_from_currency.get()
        to_currency = combo_to_currency.get()

        if from_currency == to_currency:
            converted_amount.set(amount)
        else:
            result = amount * (exchange_rates[to_currency] / exchange_rates[from_currency])
            converted_amount.set(round(result, 2))
    except ValueError:
        converted_amount.set("Invalid Input")

# Create the main window
root = tk.Tk()
root.geometry('600x600')
root.title("Currency Converter")

# Label and Entry for the amount to convert
label_amount = ttk.Label(root, text="Amount:")
label_amount.grid(row=0, column=0, padx=10, pady=10)
entry_amount = ttk.Entry(root)
entry_amount.grid(row=0, column=1, padx=10, pady=10)

# Combobox for the "From" currency
label_from_currency = ttk.Label(root, text="From Currency:")
label_from_currency.grid(row=1, column=0, padx=10, pady=10)
combo_from_currency = ttk.Combobox(root, values=list(exchange_rates.keys()))
combo_from_currency.grid(row=1, column=1, padx=10, pady=10)
combo_from_currency.set('USD')  # Default selection

# Combobox for the "To" currency
label_to_currency = ttk.Label(root, text="To Currency:")
label_to_currency.grid(row=2, column=0, padx=10, pady=10)
combo_to_currency = ttk.Combobox(root, values=list(exchange_rates.keys()))
combo_to_currency.grid(row=2, column=1, padx=10, pady=10)
combo_to_currency.set('EUR')  # Default selection

# Button to perform the conversion
convert_button = ttk.Button(root, text="Convert", command=convert_currency)
convert_button.grid(row=3, columnspan=2, padx=10, pady=10)

# Label to display the converted amount
converted_amount = tk.StringVar()
result_label = ttk.Label(root, textvariable=converted_amount)
result_label.grid(row=4, columnspan=2, padx=10, pady=10)

# Start the Tkinter main loop
root.mainloop()
```
