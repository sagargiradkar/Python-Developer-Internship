```python
import tkinter as tk
from tkinter import messagebox

# Initialize the main window
root = tk.Tk()
root.geometry('600x600')
root.title("Tic-Tac-Toe")

# Create variables to keep track of the current player and the board state
current_player = "X"
board = [""] * 9

# Function to check for a win or a draw
def check_winner():
    # Check rows, columns, and diagonals for a win
    for i in range(3):
        if board[i] == board[i + 3] == board[i + 6] != "":
            return True
        if board[i * 3] == board[i * 3 + 1] == board[i * 3 + 2] != "":
            return True
    if board[0] == board[4] == board[8] != "":
        return True
    if board[2] == board[4] == board[6] != "":
        return True

    # Check for a draw
    if "" not in board:
        messagebox.showinfo("Tic-Tac-Toe", "It's a draw!")
        root.quit()

    return False

# Function to handle a player's move
def handle_click(button, index):
    global current_player

    # Check if the square is empty
    if board[index] == "":
        board[index] = current_player
        button.config(text=current_player)
        
        if check_winner():
            messagebox.showinfo("Tic-Tac-Toe", f"Player {current_player} wins!")
            root.quit()
        
        # Switch to the other player
        current_player = "O" if current_player == "X" else "X"

# Create buttons for the Tic-Tac-Toe grid
buttons = []
for i in range(3):
    for j in range(3):
        button = tk.Button(root, text="", font=("normal", 20), width=6, height=2,
                           command=lambda i=i, j=j: handle_click(buttons[i * 3 + j], i * 3 + j))
        button.grid(row=i, column=j)
        buttons.append(button)

# Start the main loop
root.mainloop()
```
