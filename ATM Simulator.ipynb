```python
import tkinter as tk
from tkinter import messagebox

# Initialize account balance
balance = 1000

# Function to deposit money
def deposit():
    global balance
    amount = float(deposit_entry.get())
    balance += amount
    update_balance_label()

# Function to withdraw money
def withdraw():
    global balance
    amount = float(withdraw_entry.get())
    if amount <= balance:
        balance -= amount
        update_balance_label()
    else:
        messagebox.showerror("Error", "Insufficient balance!")

# Function to update balance label
def update_balance_label():
    balance_label.config(text="Balance: ${:.2f}".format(balance))

# Create the main window
root = tk.Tk()
root.geometry('600x600')
root.title("ATM Simulator")

# Create and configure widgets
deposit_label = tk.Label(root, text="Deposit:")
deposit_entry = tk.Entry(root)
deposit_button = tk.Button(root, text="Deposit", command=deposit)

withdraw_label = tk.Label(root, text="Withdraw:")
withdraw_entry = tk.Entry(root)
withdraw_button = tk.Button(root, text="Withdraw", command=withdraw)

balance_label = tk.Label(root, text="Balance: ${:.2f}".format(balance))

# Place widgets on the window
deposit_label.grid(row=0, column=0)
deposit_entry.grid(row=0, column=1)
deposit_button.grid(row=0, column=2)

withdraw_label.grid(row=1, column=0)
withdraw_entry.grid(row=1, column=1)
withdraw_button.grid(row=1, column=2)

balance_label.grid(row=2, column=0, columnspan=3)

# Start the Tkinter main loop
root.mainloop()
```
